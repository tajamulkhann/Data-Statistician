```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\eli\\Desktop\\Piece_Dim.csv")
df.head()
```

```python
sns.relplot(data = df, x = "Length", y = "Width")
plt.show()
```

```python
sns.relplot(data = df, x = "Length", y = "Width", hue = "Operator")
plt.show()
```

```python
sns.relplot(data = df, x = "Length", y = "Width", col = "Operator", marker = "D")
plt.show()
```

```python
sns.relplot(data = df, x = "Length", y = "Width", row = "Operator")
plt.show()
```

```python
sns.relplot(data = df, x = "Item_No", y = "Width", kind = "line", aspect = 3)
mean = np.mean(df["Width"])
std = np.std(df["Width"])
plt.axhline(y = mean, color = "red", linestyle = "-")
plt.axhline(y = (mean + 3 * std), color = "red", linestyle = "-")
plt.axhline(y = (mean - 3 * std), color = "red", linestyle = "-")
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.regplot(data = df, x = "Length", y = "Width")
plt.show()
```
