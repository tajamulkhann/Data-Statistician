```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\eli\\Desktop\\Piece_Dim.csv")
df.head()
```

```python
mean_l = np.mean(df["Length"])
std_l = np.std(df["Length"])
```

```python
plt.plot(df["Length"])
plt.axhline(y = mean_l, color = "red", linestyle = "--")
plt.axhline(y = (mean_l + 3 * std_l), color = "red", linestyle = "--")
plt.axhline(y = (mean_l - 3 * std_l), color = "red", linestyle = "--")
plt.title("Length")
plt.show()
```

```python
plt.scatter(df["Length"], df["Height"], marker = "D")
plt.title("Relationship between Length and Height")
plt.show()
```
